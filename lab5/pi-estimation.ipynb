# How it works ?



## Methodology
Monte Carlo integration technique:

    - Defining the Problem Domain: The goal is to approximate the value of pi (π) using a Monte Carlo simulation. The idea is to inscribe a circle within a square, where the diameter of the circle is equal to the length of the square's sides.
    - Generating Random Points: Generate a large number of random points (x, y) within the square. The points are uniformly distributed within the square.
    - Counting Points Inside the Circle: Count the number of points (x, y) that fall within the circle. This can be done by checking if the distance of each point from the center of the circle is less than or equal to the radius of the circle (which is half the length of the square's sides).
    - Calculating the Ratio: Calculate the ratio of the number of points inside the circle to the total number of points generated. This ratio represents the ratio of the area of the circle to the area of the square.
    - Approximating Pi: Since the area of the circle is given by πr^2, where r is the radius, and the area of the square is given by (2r)^2, the ratio of the areas is: Area of circle / Area of square = πr^2 / (2r)^2 = π/4 Therefore, the value of π can be approximated as: π ≈ 4 * (Number of points inside the circle) / (Total number of points generated)
    - Improving Accuracy: To improve the accuracy of the approximation, you can increase the number of random points generated. As the number of points increases, the ratio of points inside the circle to the total points will converge to the true value of π/4.


```python
import numpy as np 
```

```python
def simple_pi(times):
    n_hits = 0
    m_trials = times
    counter = 0 
    
    while counter < m_trials:
        x = np.random.uniform(0,1)
        y = np.random.uniform(0,1)
        if x**2 + y**2 < 1:
            n_hits += 1
        counter += 1
    
    return 4 * n_hits / m_trials

simple_pi(1e5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_error(max_times=5):
    error = []
    times = [i * 10 for i in range(1, max_times)]
    for i in times:
        error.append(abs(np.pi - simple_pi(i)))

    # Plot scatter plot
    plt.scatter(range(1, max_times), error, label='Error')

    # Set labels and title
    plt.xlabel('Number of Times')
    plt.ylabel('Error')
    plt.title('Error vs Number of Times')

    # Add the line with slope -1/2
    x = np.array(range(1, max_times))
    y = 1 / np.sqrt(x)
    plt.plot(x, y, label='Slope=-1/2', color='r')

    # Show legend
    plt.legend()

    # Show the plot
    plt.show()

plot_error(100)
```
