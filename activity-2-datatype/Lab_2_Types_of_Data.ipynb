####List Form

```python
import pandas as pd

data = pd.read_excel('/content/Facebook_Data.xlsx')

users = data.columns.tolist()[1:]
connections_dict = {}
n=10

for i, user1 in enumerate(users[:n]):
    connections = []
    for j, user2 in enumerate(users):
        if i != j and data.loc[i, user2] == 1:
            connections.append(user2)
    connections_dict[user1] = connections

# Display connections for each user
print("Connections:")
index=1
for user, connections in connections_dict.items():
    print(f"{index}.{user}: {', '.join(connections)}")
    index=index+1
```

####Graph Representation

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

n=100

data = pd.read_excel('/content/Facebook_Data.xlsx')

users = data.columns.tolist()[1:(n+1)]
G = nx.Graph()

for i, user1 in enumerate(users):
    for j, user2 in enumerate(users):
        if i != j and data.loc[i, user2] == 1:
            G.add_edge(user1, user2)

plt.figure(figsize=(12, 8))
nx.draw(G, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_color='black')
plt.title(f"Social Media Network Graph (First {n} Users)")
plt.show()
```

ข้อมูลที่นำเข้าคือ Facebook Social Network Data ซึ่งข้อมูลนี้เป็น Graph/Network Data เพราะข้อมูลในแต่ละแถวหรือคอลัมน์นั้นมีความการเชื่อมต่อกัน โดยให้ Attributes เป็น Node ภายในกราฟ และค่า 1 แทนเส้นว่าระหว่าง Node นั้นมี Edge เชื่อมกัน และค่า 0 คือไม่มี Edge เชื่อมกัน โดยที่ Attributes คือ username โดยสามารถแสดงข้อมูลได้ในรูปแบบ List, Matrix, Graph ตามข้างบน

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

n=100
data_subset = data.head(n)

cross_table = pd.crosstab(data_subset['Gender'], data_subset['Company'])

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_table, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
plt.title('Car Sales by Gender and Company (First 100 Rows)')
plt.xlabel('Company')
plt.ylabel('Gender')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

ข้อมูลแถวเป็นเพศ (Gender) และข้อมูลคอลัมน์เป็นยี่ห้อรถที่ขายได้ ซึ่งข้อมูลยอดซื้อรถในคอลัมน์เพิ่มขึ้นด้วยเวลาหรือเหตุการณ์ , แสดงให้เห็นถึงการเปลี่ยนแปลงของ attributes ในระยะหรือเหตุการณ์ต่าง ๆ ในที่นี้คือความสัมพันธ์ระหว่างเพศของลูกค้ากับยอดการซื้อขายรถยนต์ และเป็นข้อมูลรายงาน (summary report) สรุปยอดการขายรถยนต์ยี่ห้อต่าง ๆ ซึ่งเป็นปกติทั่วไปที่ใช้ Crosstable Data ในการทำข้อมูล report

เป็นการอ่านข้อมูล JSON เข้าไปในแบบ data frame ชื่อ drake (ชื่อศิลปินสากล) เป็นข้อมูลประเภท semi-structure data เนื่องจาก
- ข้อมูล Key-Value Pairs: แต่ละแถวของตารางมีข้อมูลในรูปแบบของคู่ค่า (Key-Value pairs) ที่กำหนดโดยคอลัมน์ เช่น "album" เป็น Key และ "Certified Lover Boy" เป็น Value
- ข้อมูล Array: บางคอลัมน์เก็บข้อมูลที่มีโครงสร้างแบบอาร์เรย์ (Arrays) เช่น "lyrics" ที่มีเนื้อเพลงทั้งหมดในรูปแบบของข้อความ
- ไม่มีการกำหนดโครงสร้างล่วงหน้า: ข้อมูลในตารางไม่มีการบังคับให้กำหนดโครงสร้างล่วงหน้า ซึ่งทำให้ข้อมูลมีความยืดหยุ่นและสามารถเปลี่ยนแปลงได้ตามความต้องการของข้อมูล

```python
covid_df = pd.read_json('https://covid19.ddc.moph.go.th/api/Cases/today-cases-line-lists')
covid_df.head()
```

เป็นการอ่านข้อมูล JSON เข้าไปในแบบ data frame covid_df เป็น semi-structured data เนื่องจากมีลักษณะที่ไม่ต้องการกำหนดโครงสร้างล่วงหน้าอย่างแน่นอน
- Key-Value Pairs: แต่ละแถวของตารางมีคู่ค่าที่กำหนดโดยชื่อคอลัมน์ (Key) และค่าที่เกี่ยวข้อง (Value). เช่น, "year" คือ Key และ 2024 คือ Value
- Array (อาเรย์): บางคอลัมน์เป็นข้อมูลที่เก็บเป็นกลุ่มหรือลำดับของข้อมูล เช่น "region_odpc" ระบุภูมิภาคที่เกี่ยวข้องกับผู้ป่วยและอาจมีค่ามากกว่าหนึ่ง

### 2.2 HTML (Web Scraping)

```python
import requests
from bs4 import BeautifulSoup

# URL of the website to scrape
url = 'https://www.artificial-intelligence.blog/'
# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')
    for link in links:
        print(link.get('href'))

    headings = soup.find_all('h1')
    for heading in headings:
        print(heading.text)
else:
    print('Failed to retrieve the webpage')
```

```python
import os
import random
from IPython.display import Image, display

folder_path = "/content/images/images"

files = os.listdir(folder_path)

image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png'))]

num_images_to_display = 10

selected_images = random.sample(image_files, num_images_to_display)

image_width = 200
image_height = 200

for image_file in selected_images:
    image_path = os.path.join(folder_path, image_file)
    display(Image(filename=image_path, width=image_width, height=image_height))
```

```python
import os
import random
import matplotlib.pyplot as plt
from skimage import io
from skimage.transform import resize

folder_path = "/content/images/images"

files = os.listdir(folder_path)

image_files = [file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png'))]

num_images_to_display = 10

selected_images = random.sample(image_files, num_images_to_display)

image_width = 200
image_height = 200

for image_file in selected_images:
    image_path = os.path.join(folder_path, image_file)
    image = io.imread(image_path)
    resized_image = resize(image, (image_height, image_width))
    plt.imshow(resized_image)
    plt.axis('off')
    plt.show()
```
